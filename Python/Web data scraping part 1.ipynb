** Collect weather information from web API **

I used wunderground to obtain weather information for five cities during the one-month period of 9/01/17 - 9/30/17.

```python
import requests
api = '0095a615c4015055'

global days
days = []

def sept_weather_api(city):
    day_st = 20170900
    global days 
    days = []
    temp = []
    
    while day_st < 20170930:
        day_st += 1
        day_type = str(day_st)
        days.append(day_type)
        
    for day in days:
        r = requests.get('http://api.wunderground.com/api/' + api + '/history_'
                       + day + '/q/' + city + '.json')
        city_data = r.json()['history']['dailysummary']
        for i in city_data:
            day = i['date']['mday']
            day_temp = i['meantempi']
            temp.append(day_temp)
    return temp
```

```python
#Now you can call the function to return the monthly weather list for each city below
sf = sept_weather_api('CA/San_Francisco')
pa = sept_weather_api('CA/Palo_Alto')
ev = sept_weather_api('IL/Evanston')
chi = sept_weather_api('IL/Chicago')
nyc = sept_weather_api('NY/New_York')
```

```python
#Data summary table for the weather in five cities in September
print("               Minimum "" Average "" Maximum ")
print("San Francisco ",min(sf),"     ",round(sum(int(i) for i in sf)/30,1),"   ", max(sf))
print("Palo Alto     ",min(pa),"     ",round(sum(int(i) for i in pa)/30,1),"   ", max(pa))
print("Evanston      ",min(ev),"     ",round(sum(int(i) for i in ev)/30,1),"   ", max(ev))
print("Chicago       ",min(chi),"     ",round(sum(int(i) for i in chi)/30,1),"   ", max(chi))
print("New York City ",min(nyc),"     ",round(sum(int(i) for i in nyc)/30,1),"   ", max(nyc))
```

```python
#line chart showing monthly temperature
import matplotlib.pyplot as plt
import math

x = days

weather_sf = list(map(lambda x:int(x),sf))
weather_pa = list(map(lambda x:int(x),pa))
weather_ev = list(map(lambda x:int(x),ev))
weather_chi = list(map(lambda x:int(x),chi))
weather_nyc = list(map(lambda x:int(x),nyc))

plt.rcParams['figure.figsize'] = 15,6
line_sf = plt.plot(x, weather_sf, color = 'blue')
line_sf = plt.plot(x, weather_pa, color = 'red')
line_sf = plt.plot(x, weather_ev, color = 'yellow')
line_sf = plt.plot(x, weather_chi, color = 'green')
line_sf = plt.plot(x, weather_nyc, color = 'orange')

plt.ylabel('temperature (F)')
plt.xlabel('day of September')
plt.legend(['$San Francisco$', '$Palo Alto$', '$Evanston$', '$Chicago$','NYC'], loc='best', fontsize=10)
plt.show()
```
